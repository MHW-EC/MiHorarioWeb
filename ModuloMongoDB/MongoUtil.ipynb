```python
import pymongo
import json
import os
```

```python
def objectFromFile(archivo, enLinea = False):
    descriptor = open(archivo, "r", encoding="utf-8")
    jsonString = "".join([linea.strip("\r\n") for linea in descriptor])
    descriptor.close()
    return json.loads(jsonString)
```
